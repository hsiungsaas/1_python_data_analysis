```python
import pandas as pd
```

```python
df = pd.read_csv("data.csv")
```

```python
df.head()
```

```python
# generate a word count
word_count = {}
for i, row in df.iterrows():
    try:
        for word in row["narrative"].split(" "):
            if word not in word_count:
                word_count[word] = 0
            word_count[word] += 1
    except:
        pass
```

```python
word_tuples = [(key, word_count[key]) for key in word_count]
word_count_df = pd.DataFrame(word_tuples, columns=["word","count"])
word_count_df = word_count_df.sort("count", ascending=False)
word_count_df = word_count_df[word_count_df.word != ""]
```

```python
print word_count_df.head(100).plot(kind="bar", x="word", figsize=(20,6), rot=70, grid=False)
```

```python
# let's define a function that will plot line graph for number of reports that day that contain the word
series_data = {}
def plot_word_occurence(word, ylim=(0,15)):
    
    word = word.upper()
    
    # iterate over each day. query df to determine number of cases that day with the word
    for i, row in df.iterrows():
        
        if word in str(row["narrative"]):
            
            if row["date"] not in series_data:
                series_data[row["date"]] = 0
            
            series_data[row["date"]] += 1

    # generate a pandas time series object
    time_series = pd.Series(series_data)
    time_series.plot(figsize=(20,6), rot=70, grid=False, ylim=ylim)
    
plot_word_occurence("graffiti")
    
```
