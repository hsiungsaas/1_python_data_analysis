Let's begin by importing [PANDAS](http://pandas.pydata.org/), a high-performance data analysis library.

```python
import pandas as pd
```

PANDAS lets us load a CSV file into a data structure called a [DataFrame](http://pandas.pydata.org/pandas-docs/version/0.13.1/generated/pandas.DataFrame.html).

```python
df = pd.read_csv("data.csv")
```

Checking out the first few rows of the DataFrame helps us to confirm we're working with the right data.

```python
df.head()
```

Let's generate a histogram of the most popular words in the case narratives.

```python
# generate a word count
word_count = {}
for i, row in df.iterrows():
    try:
        for word in row["narrative"].split(" "):
            if word not in word_count:
                word_count[word] = 0
            word_count[word] += 1
    except:
        pass
    
# housekeeping
word_tuples = [(key, word_count[key]) for key in word_count]
word_count_df = pd.DataFrame(word_tuples, columns=["word","count"])
word_count_df = word_count_df.sort("count", ascending=False)
word_count_df = word_count_df[word_count_df.word != ""]

# plot
word_count_df.head(100).plot(kind="bar", x="word", figsize=(20,6), rot=70, grid=False)
```

Let's create a method, plot_word_occurence(word), that will plot the number of cases/day a word appears in the narrative.

```python
def plot_word_occurence(word, ylim=(0,15)):
    
    series_data = {}
    word = word.upper()
    
    # iterate over each day. query df to determine number of cases that day with the word
    for i, row in df.iterrows():
        
        if word in str(row["narrative"]):
            
            if row["date"] not in series_data:
                series_data[row["date"]] = 0
            
            series_data[row["date"]] += 1

    # generate a pandas time series object
    time_series = pd.Series(series_data)
    time_series.plot(figsize=(20,6), rot=70, grid=False, ylim=ylim)
```

Our method can now be recycled to plot the occurence of any word we'd like.

```python
plot_word_occurence("theft")
```

```python
plot_word_occurence("bicycle")
```

How about some seasonal trends? Let's aggregate the __data by month__.

```python
def plot_word_occurence_month(word, ylim=(0,15)):
    
    df_data = {"January":0,"February":0,"March":0,"April":0,"May":0,"June":0,"July":0,"August":0,"September":0,"October":0,"November":0,"December":0}
    word = word.upper()
    
    # iterate over each day. query df to determine number of cases that day with the word
    for i, row in df.iterrows():

        if word in str(row["narrative"]):
            
            month_of_year = datetime.datetime.strptime(row["date"],"%Y-%m-%d 00:00:00").strftime("%B")
            df_data[month_of_year] += 1

    df_tuples = [(key, df_data[key]) for key in df_data]
    word_count_df = pd.DataFrame(df_tuples, columns=["day","count"])
    
    word_count_df.sort("count", ascending=False).plot(kind="bar", x="day", figsize=(20,6), rot=70, grid=False)
```

Let's look at general theft. 

```python
plot_word_occurence_month("theft")
```

Perhaps we're more lax with our property earlier in the year?

```python
plot_word_occurence_month("bicycle")
```

Bikes seem to go pretty early in the school year.

```python
plot_word_occurence_month("alcohol")
```
